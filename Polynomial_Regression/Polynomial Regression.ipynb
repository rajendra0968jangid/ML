Polynomial Regression is a type of regression used when the relationship between input (X) and output (Y) is non-linear (curved), but can be modeled using a polynomial equation.

Although it looks nonlinear, it is still a linear regression model because it is linear in terms of coefficients.

Why Polynomial Regression?

Simple linear regression fits a straight line: Y = mX + C

But many real-life problems follow a curve:
Population growth
Temperature variation
Salary vs Experience (sometimes)
Vehicle speed vs fuel consumption
For such cases, polynomial regression is useful.

Two Types Of Polynomial:- 
1) Single Input Polynomial Regression
   Suppose one input variable x.
   Y=m1x + m2x^2 + C
   We must find:
   C:- intercept
   m1:- coefficient of x
   m2:- coefficient of x^2

   m= ((X^TX)^-1X^T)Y or Quadratic Equation to Find the Value of m1 & m2

2) Multiple Input Polynomial Regression
   General Equation of Multiple Linear Regression
   Y=m1x + m2x^2 + m3x^3 + m4x^4 + m5x^5 + ..... mnx^n + C
   Use only m= ((X^TX)^-1X^T)Y to find the value of m 



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Dataset
data = {
    'Area': [1000,1200,1500,1800,2000,2300,2500,2700],
    'Bedrooms': [2,3,3,4,4,5,5,6],
    'Age': [10,8,7,5,4,3,2,1],
    'Price': [25,30,36,45,50,60,65,72]
}
df = pd.DataFrame(data)
df
```

```python
# Input and Output
X = df[['Area', 'Bedrooms', 'Age']]
y = df['Price']
print(X)
print(y)
```

```python
# Polynomial Features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
X_poly
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_poly,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
# Model Training
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
# Prediction
y_pred = model.predict(X_test)

print("Predicted Values:")
print(y_pred)
```

```python
# Error
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```

```python
# New Prediction
new_data = [[2100, 4, 3]]

new_poly = poly.transform(new_data)

prediction = model.predict(new_poly)

print("Predicted House Price:", prediction)
```
