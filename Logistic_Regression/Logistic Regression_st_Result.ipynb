```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("st_result.csv")
#data["RESULT"] = data["RESULT"].map({"FAIL":0,"PASS":1})
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]

#X = df.drop("RESULT", axis=1)
#y = df["RESULT"]
Y
```

z=m1​x1 ​+ m2​x2 ​+m3​x3 ​+m4​x4 ​+ m5​x5 + C​

z=m1​(StudyHours)+m2​(Attendance)+m3​(Assignment)+m4​(InternalMarks)+m5​(PreviousPercetage) + C

Then Logistic Regression applies the Sigmoid Function:
p=1/(1+e−z)​

where
p = Probability of PASS
1−p = Probability of FAIL

```python
# Convert YES/NO into 1/0
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Predicted Values:")
print(y_pred)
```

```python
# Probability Prediction
y_prob = model.predict_proba(X_test)
print("\nProbability:")
print(y_prob)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy =", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
```

```python
report = classification_report(y_test, y_pred)

print(report)
```

```python
print(model.coef_)
print(model.intercept_)
```
