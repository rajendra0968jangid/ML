```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Year': [1960, 1970, 1980, 1990, 2000, 2010, 2020],
    'Population': [0.45, 0.55, 0.69, 0.85, 1.05, 1.23, 1.38]
}
```

```python
df = pd.DataFrame(data)
X = df[['Year']]  # Feature (Year)
y = df['Population']  # Target (Population)
```

```python
model = LinearRegression()
model.fit(X, y)

# Predict future population
future_years = np.array([2030, 2040, 2050]).reshape(-1, 1)
predicted_population = model.predict(future_years)

# Display predictions
for year, pop in zip(future_years, predicted_population):
    print(f"Predicted population in {year[0]}: {pop:.2f} billion")
```

```python
# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Historical data')
#plt.bar(X, y, color='red', label='Regression line')
plt.plot(future_years, predicted_population, color='green', label='Predictions')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.legend()
plt.title('India Population Prediction')
plt.show()
```
