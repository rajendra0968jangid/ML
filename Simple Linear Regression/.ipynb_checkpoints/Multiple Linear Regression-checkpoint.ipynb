Multiple Linear Regression

Multiple Linear Regression (MLR) is a supervised machine learning algorithm used to predict a continuous dependent variable using two or more independent variables.

General Equation of Multiple Linear Regression
Y=m1x1 + m2x2 + m3x3 + m4x4 + m5x5 + ..... mnxn + C

m= ((X^TX)^-1X^T)Y

Where:
Y = Dependent variable (Output)
X₁ = Study Hours
X₂ = Attendance (%)
X₃ = Assignment Score
X₄ = Internal Marks
X₅ = Practical Marks
C = Intercept
m1...m5 = Regression Coefficients

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("st_data.csv")
#data.count()[0]
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# Independent variables
X = df.iloc[:,:-1]
# Dependent variable
Y = df['FinalMarks']
print(X)
print(Y)
```

```python
# Split data
X_train,X_test,Y_train,Y_test = train_test_split(
    X,Y,test_size=0.3,random_state=80)
print(X_test)
print(Y_train)
```

```python
# Model
model = LinearRegression()
model.fit(X_train,Y_train)

# Intercept
print("Intercept:",model.intercept_)

# Coefficients
print("Coefficients:",model.coef_)
```

```python
# Prediction for new student
new_data = [[10,89,83,81,85]]
prediction = model.predict(X_test)

print("Predicted Final Marks =",prediction)
```
