Simple Linear Regression is a supervised machine learning algorithm used to predict 
a continuous numerical value using one independent variable.

Mathematical Equation:-
y = mx + c
Where:
y	Predicted Output / Dependent Variable / Target Variable / Response Variable / Output variable
x	Input variable / Independent Variable / Feature / Predictor
m	Slope (Coefficient)
c	Intercept

m=∑(xi​−xˉ)(yi​−yˉ​)​ / ∑(xi​−xˉ)2

Best Fit Line:-
Regression tries to find the line with minimum error.

Error / Residual:-
Difference between actual and predicted is called Error / Residual
Residual= Actual Value−Predicted Value
OR
e=y−y(p)

Cost Function:-
Regression minimizes error using cost function.
Mean Squared Error (MSE)
MSE= 1/n [​∑(yi​−y^​i​)2]

MAE (Mean Absolute Error)
MAE =1/n ​∑∣yi​−y^​i​∣
​
RMSE

RMSE= Root of 1/n​∑(yi​−y^​i​)2

```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
Y = np.array([50, 55, 65, 70, 80])
print(X)
print(Y)
```

```python
# Create & Train model 
model = LinearRegression()
model.fit(X, Y)
```

```python
# Get slope and intercept
m = model.coef_
b = model.intercept_

print("Slope (m):", m)
print("Intercept (C):", b)
```

```python
# Prediction value
y_pred = model.predict(X)
print(y_pred)
```

```python
# Predict new value
new_value = np.array([10,20,11,1]).reshape(-1,1)
print("Prediction for X=6:", model.predict(new_value))
```

```python
# Plot graph
plt.scatter(X, Y)
plt.plot(X, y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()
```
